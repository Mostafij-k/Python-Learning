```python
# Write a function to print the length of a list
cities=['Dhaka', 'Khulna', 'Barishal', 'Rajshai', 'Kushtia', 'Jessore']
Fruits=['Mango', 'Orange', 'Pineapple', 'Guava', 'Jackfruits']
def list_length(list):
    print(len(list))
list_length(cities)
list_length(Fruits)

# write a function to print the element of a list
def item_list(list):
    for item in list:
        print(item, end=' ')
item_list(cities)
print()
item_list(Fruits)
```
