```python
# Scope: 
# Scope means the area of the program where a variable is accessible
# There are two types:
    # 1. Global Scope
    # 2. Local Scope
# Python decides variable visibility using a rule called LEGB rule
# L: Local G: Global E: Enclosing D: Built In

# A global variable is declared outside of any function
# Example:
x=15 # X is global variable
def display():
    print("Inside variable: ", x)
display()
print("Outside Variable: ", x)
```

```python
# A local variable is defined inside a function.
# It is only accessible inside that function
# Example:
def display():
    y=10
    print('Inside function: ', y)
display()
# print('Outside function: ', y) # Give an error
    # y exists only while the function is running.
    # After the function finishes, y disappears.
    # So, print(y) outside gives NameError.
```

```python
# If you declare a variable inside a function with the same name as a global variable,
# Python will treat it as local inside that function.
x=100
def display():
    x=50
    print('Inside variable: ', x)
display()
print('Outside function: ', x)
```

```python
# To modify a global variable inside a function, tell Python using the global keyword.
count=0
def display():
    global count
    count+=1
    print('Inside variable: ', count)
display()
print('Outside variable: ', count)
```

```python
                    # ---------Enclosing Scope-----#
# Enclosing Scope refers to the scope of a function surrounds by another function.
# It allows the inner function to access variables from the outer (enclosing) function.

    # Nested function: Function inside another function
    # Outer function becomes the enclosing scope for the inner function
    # The outer function cannot access variables of the inner function

# Enclosing variable: Variable from outer function used by inner one

def outer():
    x=10
    def inner():
        print(x)
    inner()
outer()
```

```python
# Example of Enclosing
def greet():
    message='Hello'
    def say():
        return message + ', World!'
    x=say()
    print(x)
greet()
```

```python
# Hierarchy Example
x='global'
def outer():
    x='Enclosing'
    def inner():
        x='Local'
        print(x)
    inner()
    print(x)
outer()
print(x)
```
